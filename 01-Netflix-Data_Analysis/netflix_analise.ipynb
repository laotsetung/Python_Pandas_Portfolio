#GRAFICO VISUALIZAÇÃO POR PERFIL (views per account)

```python
arq["Profile Name"].value_counts()

Total_views=arq['Profile Name'].value_counts()

#cria um grafico de pizza, sobre a soma de valores de 'Profile Name'
plt.pie(arq['Profile Name'].value_counts(), autopct= lambda z1: '{:.0f}'.format(z1*Total_views.sum()/100), labels=Total_views.index, radius=1.6, textprops={'fontsize': 14})
plt.title("Visualizações na NETFLIX por Perfil:",pad=90, fontdict={'fontsize': 18})
plt.savefig('viz_por_perfil_2.png', bbox_inches='tight')
plt.show()
```

```python
#REMOVING ENTRIES WITH LESS THAN 5.MINUTES OF VIEW DURATION
arq1 = arq.loc[arq["Duration"] > '00:05:00']

arq1.info()
```

```python
#SEPARATE DF INTO 5 NEW DF´S FOR EACH USER ACCOUNT
perfil1 = arq1.loc[arq['Profile Name'] == 'Perfil 1']
perfil2 = arq1.loc[arq['Profile Name'] == 'Perfil 2']
perfil3 = arq1.loc[arq['Profile Name'] == 'Perfil 3']
perfil4 = arq1.loc[arq['Profile Name'] == 'Perfil 4']
perfil5 = arq1.loc[arq['Profile Name'] == 'Perfil 5']

perfil1.info()
```

#USING PERFIL1 TO GENERATE MORE GRAPHCS

```python
#DIVIDE THE DATE CELL INTO 2 NEW CELLS (SPLIT YEAR AND MONTH)
ano = perfil1["Start Time"].str.split(' ', expand=True)[0].str.split('-', expand=True)[0]
mes = perfil1["Start Time"].str.split(' ', expand=True)[0].str.split('-', expand=True)[1]

perfil1 = perfil1.assign(Data_ano=ano)
perfil1 = perfil1.assign(Data_mes=mes)
perfil1
```

#GRAPHCS FOR 'PERFIL 1'

```python
#VISUALIZATION OF VIEWS IN PERFIL 1 BY YEAR 
values=perfil1['Data_ano'].value_counts()

#autopct= (lambda function to generate labels values)
plt.pie(perfil1['Data_ano'].value_counts(), autopct= lambda z1: '{:.0f}'.format(z1*values.sum()/100), labels=values.index, radius=1.6, textprops={'fontsize': 14})
plt.title("Visualizações na NETFLIX por ano: Perfil 1",pad=90, fontdict={'fontsize': 18})
plt.savefig('perfil1_vis_por_ano.png', bbox_inches='tight')
plt.show()
```

```python
#VIEWS OF 2017 PER MONTH
perfil1_17 = perfil1[perfil1['Data_ano'] == '2017']
perfil1_17 = perfil1_17['Data_mes'].value_counts()

perfil1_17 = perfil1_17.sort_index(0)

plt.bar(perfil1_17.index, perfil1_17)
plt.title("Visualização por mês do ano de 2017: Perfil 1")
plt.grid(True)
plt.savefig('ano2017_perfil1.png', bbox_inches='tight')
plt.show()
```
